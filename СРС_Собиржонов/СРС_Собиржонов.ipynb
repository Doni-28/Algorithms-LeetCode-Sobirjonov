```python
import time
import matplotlib.pyplot as plt
import random


# Пузырьковая сортировка
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        swapped = False
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                swapped = True
        if not swapped:
            break
    return arr

# Генерация массива
def generate_array(size, type="random"):
    if type == "random":
        return [random.randint(0, 10000) for _ in range(size)]
    elif type == "sorted":
        return list(range(size))
    elif type == "reversed":
        return list(range(size, 0, -1))
    elif type == "almost":
        arr = list(range(size))
        for _ in range(size // 20):
            i, j = random.randint(0, size-1), random.randint(0, size-1)
            arr[i], arr[j] = arr[j], arr[i]
        return arr
    else:
        raise ValueError("Неизвестный тип массива")


# Настройки теста
sizes = [1000, 5000, 10000, 50000]
times = []


# Замер времени
for n in sizes:
    arr = generate_array(n, type="random")  #  меняет тип массива
    start = time.time()
    bubble_sort(arr)
    end = time.time()
    execution_time = end - start
    times.append(execution_time)

# Построение графика
plt.plot(sizes, times, marker='o', label="Bubble Sort")

plt.xlabel("Размер входных данных (n)")
plt.ylabel("Время выполнения (секунды)")
plt.title("Время выполнения Bubble Sort")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import random
import time
import matplotlib.pyplot as plt

# -----------------------------
# Реализация QuickSort с Ломуто
# -----------------------------
def partition(arr, low, high):
    pivot = arr[high]
    i = low - 1
    for j in range(low, high):
        if arr[j] <= pivot:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]
    arr[i + 1], arr[high] = arr[high], arr[i + 1]
    return i + 1

def quick_sort(arr, low=None, high=None):
    if low is None or high is None:
        low, high = 0, len(arr)-1
    if low < high:
        pi = partition(arr, low, high)
        quick_sort(arr, low, pi-1)
        quick_sort(arr, pi+1, high)
    return arr

# -----------------------------
# Генерация случайного массива
# -----------------------------
def generate_array(size):
    return [random.randint(0, 100000) for _ in range(size)]

# -----------------------------
# Список размеров массивов и подготовка для графика
# -----------------------------
sizes = [1000, 5000, 10000, 50000]
times = []

# -----------------------------
# Замер времени выполнения
# -----------------------------
for n in sizes:
    arr = generate_array(n)
    start = time.time()
    quick_sort(arr)
    end = time.time()
    execution_time = end - start
    times.append(execution_time)

# -----------------------------
# Построение графика
# -----------------------------
plt.plot(sizes, times, marker='o', color='red')
plt.xlabel("Размер массива")
plt.ylabel("Время выполнения (сек)")
plt.title("QuickSort (Lomuto Partition): время выполнения")
plt.grid(True)
plt.show()
```

```python
import time
import matplotlib.pyplot as plt
import random

# Bubble Sort O(n²)
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        swapped = False
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                swapped = True
        if not swapped:
            break
    return arr

# QuickSort O(n log n) (Ломуто)
def partition(arr, low, high):
    pivot = arr[high]
    i = low - 1
    for j in range(low, high):
        if arr[j] <= pivot:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]
    arr[i+1], arr[high] = arr[high], arr[i+1]
    return i + 1

def quick_sort(arr, low=None, high=None):
    if low is None or high is None:
        low, high = 0, len(arr)-1
    if low < high:
        pi = partition(arr, low, high)
        quick_sort(arr, low, pi-1)
        quick_sort(arr, pi+1, high)
    return arr

# Генерация случайного массива
def generate_array(size):
    return [random.randint(0, 10000) for _ in range(size)]

# Настройки теста
sizes = [1000, 5000, 10000, 50000]
times_bubble = []
times_quick = []

# Замер времени
for n in sizes:
    arr = generate_array(n)

    # Bubble Sort
    start = time.time()
    bubble_sort(arr.copy())
    end = time.time()
    times_bubble.append(end - start)

    # QuickSort
    start = time.time()
    quick_sort(arr.copy())
    end = time.time()
    times_quick.append(end - start)

# Построение графика
plt.plot(sizes, times_bubble, marker='o', label='Bubble Sort O(n²)')
plt.plot(sizes, times_quick, marker='o', label='QuickSort O(n log n)')

plt.xlabel("Размер входных данных (n)")
plt.ylabel("Время выполнения (секунды)")
plt.title("Сравнение времени выполнения алгоритмов сортировки")

plt.legend()
plt.grid(True)
plt.show()
```
